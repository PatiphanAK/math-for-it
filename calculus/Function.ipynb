# Functions, Graphs, and Models

## Function and Their Graph
Defination : A function f from a set D to a set Y is a rule that assigns a single
value f x( ) in Y to each x in D.

```python
import numpy as np
import matplotlib.pyplot as plt
```

### example graph of function
$$y = {x+2}$$

```python
x = np.linspace(-5, 5, 10)
y = x + 2
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='y = x+2', color='blue', linestyle='-')
plt.title('Graph of y = x + 2')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.show()
```
