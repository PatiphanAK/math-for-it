## Basic Rules of Derivative

### Derivative of a Constant
The derivative of a constant function is always zero. This can be understood intuitively and is also proven by the limit definition of a derivative.

Let's consider a function $f(x)$ where the output is a constant value, $c$, regardless of the input $x$. In other words, $f(x) = c$. The graph of this function is a horizontal line.

Using the **limit definition** of the derivative, we have:

$$f'(x) = \lim_{h \to 0} \frac{f(x+h) - f(x)}{h}$$

Since $f(x) = c$ for all values of $x$, both $f(x+h)$ and $f(x)$ are equal to $c$. Substituting these into the formula:

$$f'(x) = \lim_{h \to 0} \frac{c - c}{h} = \lim_{h \to 0} \frac{0}{h}$$

As $h$ approaches 0, the numerator is always 0, so the entire expression is 0.

$$f'(x) = 0$$

In Leibniz notation, this is written as:

<div style="border: 6px solid #0ea5e9; background-color:#f0f9ff; padding: 10px; margin:10px 0;">
$$\frac{d}{dx}(c) = 0$$
</div>

This result makes sense geometrically: the derivative represents the slope of the tangent line to the function's graph. A constant function is a horizontal line, and the slope of any horizontal line is always 0.

```python
# Define the constant function and its derivative
constant_function = lambda x: 5 * np.ones_like(x)
derivative_of_constant = lambda x: 0 * np.ones_like(x)

# Create x values for plotting
x_values = np.linspace(-10, 10, 400)
y_constant = constant_function(x_values)
y_derivative = derivative_of_constant(x_values)

# Choose specific points to demonstrate x, x+h, f(x), f(x+h)
x_point = 2
h = 3
x_plus_h = x_point + h

f_x = constant_function(np.array([x_point]))[0]
f_x_plus_h = constant_function(np.array([x_plus_h]))[0]

# Create the plot
plt.figure(figsize=(12, 8))

# Plot the constant function and its derivative
plt.plot(x_values, y_constant, label='$f(x) = 5$ (Constant Function)', color='blue', linestyle='-', linewidth=2)
plt.plot(x_values, y_derivative, label="$f'(x) = 0$ (Derivative)", color='red', linestyle='--', linewidth=2)

# Mark the specific points x and x+h
plt.plot(x_point, f_x, 'go', markersize=10, label=f'$f({x_point}) = {f_x}$')
plt.plot(x_plus_h, f_x_plus_h, 'mo', markersize=10, label=f'$f({x_plus_h}) = {f_x_plus_h}$')

# Draw vertical lines from x-axis to the function points
plt.axvline(x_point, color='green', linestyle=':', alpha=0.7, linewidth=1.5)
plt.axvline(x_plus_h, color='magenta', linestyle=':', alpha=0.7, linewidth=1.5)

# Draw horizontal lines to show f(x) and f(x+h) values
plt.axhline(f_x, color='green', linestyle=':', alpha=0.5, linewidth=1)
plt.axhline(f_x_plus_h, color='magenta', linestyle=':', alpha=0.5, linewidth=1)

# Draw a line connecting the two points to show the slope
plt.plot([x_point, x_plus_h], [f_x, f_x_plus_h], 'k--', alpha=0.7, linewidth=2, label='Secant Line')

# Add annotations for the points
plt.annotate(f'x = {x_point}', xy=(x_point, -1), xytext=(x_point, -1.5),
             ha='center', fontsize=12, color='green',
             arrowprops=dict(arrowstyle='->', color='green', alpha=0.7))

plt.annotate(f'x + h = {x_plus_h}\n(h = {h})', xy=(x_plus_h, -1), xytext=(x_plus_h, -1.5),
             ha='center', fontsize=12, color='magenta',
             arrowprops=dict(arrowstyle='->', color='magenta', alpha=0.7))

plt.annotate(f'f(x) = {f_x}', xy=(x_point, f_x), xytext=(x_point - 2, f_x + 0.8),
             ha='center', fontsize=10, color='green',
             arrowprops=dict(arrowstyle='->', color='green', alpha=0.7))

plt.annotate(f'f(x+h) = {f_x_plus_h}', xy=(x_plus_h, f_x_plus_h), xytext=(x_plus_h + 2, f_x_plus_h + 0.8),
             ha='center', fontsize=10, color='magenta',
             arrowprops=dict(arrowstyle='->', color='magenta', alpha=0.7))

# Add text box showing the derivative calculation
slope = (f_x_plus_h - f_x) / (x_plus_h - x_point)
textstr = f'''Derivative Calculation:
$\\frac{{f(x+h) - f(x)}}{{h}} = \\frac{{{f_x_plus_h} - {f_x}}}{{{h}}} = \\frac{{0}}{{{h}}} = 0$

Since f(x) is constant, f(x+h) = f(x) for any h
Therefore, the slope is always 0'''

props = dict(boxstyle='round', facecolor='wheat', alpha=0.8)
plt.text(0.02, 0.98, textstr, transform=plt.gca().transAxes, fontsize=11,
         verticalalignment='top', bbox=props)

# Set up the plot
plt.title('Constant Function: Visualizing Why the Derivative is Zero', fontsize=16, fontweight='bold')
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)
plt.axhline(0, color='gray', linewidth=0.8, linestyle=':')
plt.axvline(0, color='gray', linewidth=0.8, linestyle=':')
plt.legend(fontsize=11, loc='lower left')
plt.grid(True, linestyle=':', alpha=0.7)
plt.ylim(-2, 7)
plt.xlim(-10, 10)

# Add minor ticks for better readability
plt.minorticks_on()
plt.tight_layout()
plt.show()
```

### Derivative of $x^n$

The derivative of the function $f(x) = x^n$ can be found using the **limit definition of the derivative** for $x > 0$ and $n \in \mathbb{R}$.

Using the limit definition:

$$
f'(x) = \lim_{h \to 0} \frac{f(x+h) - f(x)}{h}
$$

Substitute $f(x) = x^n$:

$$
f'(x) = \lim_{h \to 0} \frac{(x+h)^n - x^n}{h}
$$

For positive integers $n \in \mathbb{Z}^+$, we can expand the difference of powers using the algebraic identity:

$$
(x+h)^n - x^n = h \Big((x+h)^{n-1} + (x+h)^{n-2}x + \dots + x^{n-1}\Big)
$$

Substitute this back into the limit:

$$
f'(x) = \lim_{h \to 0} \frac{h \big((x+h)^{n-1} + (x+h)^{n-2}x + \dots + x^{n-1}\big)}{h}
$$

Cancel $h$:

$$
f'(x) = \lim_{h \to 0} \big((x+h)^{n-1} + (x+h)^{n-2}x + \dots + x^{n-1}\big)
$$

As $h \to 0$, each term simplifies to $x^{n-1}$, and there are $n$ terms:

$$
f'(x) = x^{n-1} + x^{n-1} + \dots + x^{n-1} \quad (\text{n terms}) = n x^{n-1}
$$

For any real exponent $n \in \mathbb{R}$, we can define $x^n = e^{n \ln x}$. Then by the **Chain rule**:

$$
f'(x) = \frac{d}{dx} e^{n \ln x} = e^{n \ln x} \cdot \frac{d}{dx}(n \ln x) = x^n \cdot \frac{n}{x} = n x^{n-1}
$$

<div style="border: 6px solid #0ea5e9; background-color:#f0f9ff; padding: 10px; margin:10px 0;">
\[
\frac{d}{dx}\left( x^n \right) = n x^{n-1}, \quad n \in \mathbb{R}, \; x > 0
\]
</div>

This result represents the slope of the tangent line to the graph of $f(x) = x^n$.

```python
# Define the power function and its derivative
n = 3
f = lambda x: x**n
f_prime = lambda x: n * x**(n - 1)

# Choose specific points for the secant line demonstration
x_point = 1.0
h = 1.0
x_plus_h = x_point + h

# Calculate function values at these points
f_x = f(x_point)
f_x_plus_h = f(x_plus_h)
y_point = f_prime(x_point)

# Create x values for plotting
x_values = np.linspace(-2, 2, 400)
y_function = f(x_values)
y_derivative = f_prime(x_values)

# Calculate tangent line details
tangent_slope = f_prime(x_point)
# Tangent line equation: y = m(x - x1) + y1
y_tangent = tangent_slope * (x_values - x_point) + f_x

# Create the plot with a pastel color theme
plt.figure(figsize=(10, 7))

# Plot the function and its derivative
plt.plot(x_values, y_function, label='$f(x) = x^3$', color='#87CEEB', linewidth=2)  # SkyBlue
plt.plot(x_values, y_derivative, label="$f'(x) = 3x^2$", color='#FFB6C1', linestyle='--', linewidth=2)  # LightPink

# Mark the specific points
plt.plot(x_point, f_x, 'o', markersize=10, color='#6A5ACD', label=f'Point of Tangency: ({x_point}, {f_x})')  # SlateBlue
plt.plot(x_point, y_point, 'o', markersize=10, color='#fc9df0', label=f'Tangent Value: ({x_point}, {y_point})') 
plt.plot(x_plus_h, f_x_plus_h, 'o', markersize=10, color='#6A5ACD', label=f'Secant Point: ({x_plus_h}, {f_x_plus_h})')  # SlateBlue

# Draw vertical lines to indicate x and x+h
plt.axvline(x=x_point, color='#DA70D6', linestyle=':', linewidth=1.5, alpha=0.8) # Orchid
plt.axvline(x=x_plus_h, color='#DA70D6', linestyle=':', linewidth=1.5, alpha=0.8) # Orchid

# Draw the secant line
plt.plot([x_point, x_plus_h], [f_x, f_x_plus_h], '--', color='#A9A9A9', linewidth=2, label='Secant Line') # DarkGray

# Draw the tangent line
plt.plot(x_values, y_tangent, linestyle='-', color='#800080', linewidth=2, label='Tangent Line at x=1') # Purple

# Add annotations
plt.annotate(f'x = {x_point}', xy=(x_point, -1.5), xytext=(x_point, -2.5),
             ha='center', fontsize=12, color='#6A5ACD',
             arrowprops=dict(arrowstyle='->', color='#6A5ACD', alpha=0.8))

plt.annotate(f'x+h = {x_plus_h}', xy=(x_plus_h, -1.5), xytext=(x_plus_h, -2.5),
             ha='center', fontsize=12, color='#6A5ACD',
             arrowprops=dict(arrowstyle='->', color='#6A5ACD', alpha=0.8))

# Add a text box with the derivative calculation for the secant line
secant_slope = (f_x_plus_h - f_x) / h
textstr = f'''Average Rate of Change (Secant Slope):
$\\frac{{f(x+h) - f(x)}}{{h}} = \\frac{{{f_x_plus_h} - {f_x}}}{{{h}}} = {secant_slope}$

Instantaneous Rate of Change (Tangent Slope):
$f'({x_point}) = {f_prime(x_point)}$
'''
props = dict(boxstyle='round', facecolor='#F0E68C', alpha=0.7)  # Khaki
plt.text(0.03, 0.97, textstr, transform=plt.gca().transAxes, fontsize=12,
         verticalalignment='top', bbox=props)

# Set plot details
plt.title('Secant and Tangent Lines for a Power Function', fontsize=16)
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)
plt.axhline(0, color='gray', linewidth=0.8, linestyle=':')
plt.axvline(0, color='gray', linewidth=0.8, linestyle=':')
plt.legend(fontsize=11)
plt.grid(True, linestyle=':', alpha=0.7)
plt.ylim(-3, 10)
plt.xlim(-2, 2)

plt.tight_layout()
```

> This graph shows the power function $f(x) = x^3$ (solid light blue) and its derivative $f'(x) = 3x^2$ (dashed light pink), highlighting the difference between instantaneous and average rate of change.
>
> * **Secant line:** Dashed gray line connecting $(1,1)$ and $(2,8)$, slope = 7 (average rate of change over the interval).
> * **Tangent line:** At $x=1$, the instantaneous rate of change is $f'(1) = 3(1)^2 = 3$, giving the slope of the tangent line.
> * **Insight:** As $h \to 0$, the secant slope approaches the derivative. The derivative is always positive except at $x=0$, reflecting the slope of $f(x)$.
