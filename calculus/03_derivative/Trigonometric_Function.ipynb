```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

# Derivative of Sine Function

We know that the sine function is 
$$ sine(\theta) = \frac{opposite}{hypotenuse}$$

to calculate the derivative of sine function, we can combine the
$$
\begin{align}
\frac{d}{dx} \sin(x) 
&= \lim_{h \to 0} \frac{\sin(x + h) - \sin(x)}{h} \\
&= \lim_{h \to 0} \frac{\sin x \cos h + \cos x \sin h - \sin x}{h} \quad \text{(using } \sin(x+h) = \sin x \cos h + \cos x \sin h\text{)} \\
&= \lim_{h \to 0} \frac{\sin x (\cos h - 1) + \cos x \sin h}{h} \\
&= \lim_{h \to 0} \left[ \sin x \cdot \frac{\cos h - 1}{h} + \cos x \cdot \frac{\sin h}{h} \right] \\
&= \sin x \cdot \lim_{h \to 0} \frac{\cos h - 1}{h} + \cos x \cdot \lim_{h \to 0} \frac{\sin h}{h} \\
&= \sin x \cdot 0 + \cos x \cdot 1 \\
&= \cos x
\end{align}
$$

Final result:

$$
\boxed{\frac{d}{dx} \sin(x) = \cos(x)}
$$

### ðŸ“Œ Problem:
**Find the derivative of** $f(x) = \sin(x)$, **verify it numerically at** $x = \frac{\pi}{4}$, **and plot both the function and its derivative over** $[0, 2\pi]$.

```python
# --- Symbolic part ---
x = sp.symbols('x')
f = sp.sin(x)
f_prime = sp.diff(f, x)

print("Symbolic function:      ", f)
print("Symbolic derivative:    ", f_prime)
# --- Numerical verification at x = pi/4 ---
x0 = np.pi / 4
h = 1e-6

# Numerical derivative (forward difference)
numerical_deriv = (np.sin(x0 + h) - np.sin(x0)) / h
exact_deriv = np.cos(x0)

print(f"\nAt x = Ï€/4 â‰ˆ {x0:.4f}:")
print(f"Numerical derivative:   {numerical_deriv:.6f}")
print(f"Exact derivative (cos): {exact_deriv:.6f}")
print(f"Error:                  {abs(numerical_deriv - exact_deriv):.2e}")

# --- Plotting ---
x_vals = np.linspace(0, 2*np.pi, 400)
y_sin = np.sin(x_vals)
y_cos = np.cos(x_vals)

plt.figure(figsize=(10, 5))
plt.plot(x_vals, y_sin, label=r'$f(x) = \sin(x)$', color='blue')
plt.plot(x_vals, y_cos, label=r"$f'(x) = \cos(x)$", color='red', linestyle='--')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title('Sine Function and Its Derivative')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True, linestyle=':', alpha=0.7)
plt.tight_layout()
plt.show()
```
