# Rage of Change and Tangent Line
The rate of change is a fundamental concept in physics, often seen in everyday life as the rate of change of speed, which is **acceleration**.

## Average Speed

When we measure speed over a time interval, we compare the position of an object at one moment in time to its position at another moment. For example, if a falling object begins to drop at **14:00:00** and we observe its position a few seconds later (**14:00:10**), the **average speed** is calculated by dividing the change in distance by the change in time.

We can express this as a function:

$$
\text{Average Speed}(t_1, t_2) = \frac{f(t_2) - f(t_1)}{t_2 - t_1},
$$

where $f(t)$ represents the position of the object at time $t$.

If we assume the motion starts at $t_1 = x$ and ends at $t_2 = x+h$, then the formula becomes

$$
\text{Average Speed}(x, x+h) = \frac{f(x+h) - f(x)}{(x+h) - x}.
$$

Simplifying the denominator gives

$$
\text{Average Speed}(x, x+h) = \frac{f(x+h) - f(x)}{h}.
$$

This form is the **difference quotient**, and as $h \to 0$, it leads directly to the definition of the **instantaneous speed** (derivative).

### <span style="color:#ff6f3c;">Example of Average Speed of Falling Object</span>
In this example, we consider the motion of an object falling freely under gravity, ignoring air resistance.  
The position of the object at time $t$ is modeled by the function
$$
f(t) = \tfrac{1}{2} g t^2,
$$
Where $g$ is the gravitational acceleration.  
To approximate the velocity of the object, we first compute the **average speed** over an interval $[x, x+h]$, which corresponds to the slope of the secant line connecting the two points $(x, f(x))$ and $(x+h, f(x+h))$ on the curve.


Given
- Initial time $t_1$: $t_1 = 2$ s
- Final time $t_2$: $t_2 = 3$ s (This is often represented as $t + \Delta t$, where $t=2$ s and $\Delta t = 1$ s).
- Gravitational acceleration $g$: $g = 9.8 \text{ m/s}^2$

```python
# Define position function for a falling object under gravity (ignoring air resistance)
g = 9.8
f = lambda t: 0.5 * g * t**2

# Time range
t = np.linspace(1.75, 3.25, 100)
y = f(t)

# Pick a point x and interval h for difference quotient
x = 2   # seconds
h = 1   # seconds

# Average speed between x and x+h
avg_speed = (f(x+h) - f(x)) / h

# Values for secant line
secant_t = np.array([x, x+h])
secant_y = np.array([f(x), f(x+h)])

# Plot
plt.figure(figsize=(8,6))
plt.plot(t, y, label=r"$f(t) = \frac{1}{2} g t^2$", color="dodgerblue")
plt.scatter([x, x+h], [f(x), f(x+h)], color="red", zorder=5, label="Points (x, f(x)) & (x+h, f(x+h))")
plt.plot(secant_t, secant_y, "--", color="palegreen", label=f"Secant line (avg speed  {avg_speed:.2f} m/s)")

# Mark h interval on x-axis
plt.hlines(f(x), x, x+h, colors="gray", linestyles="dotted")
plt.vlines(x, 0, f(x), colors="gray", linestyles="dotted")
plt.vlines(x+h, 0, f(x+h), colors="gray", linestyles="dotted")

plt.title("Falling Object and Average Speed (Secant Line)")
plt.xlabel("Time (s)")
plt.ylabel("Position f(t) (m)")
plt.legend()
plt.grid(True)
plt.show()
```

<b>Description:</b><br>
The <b><span style="color:dodgerblue;">blue curve</span></b> shows the position of a falling object over time.<br>
The <b><span style="color:red;">red points</span></b> mark the positions at <b>t = 2 s</b> and <b>t = 3 s</b>.<br>
The <b><span style="color:green;">green dashed line</span></b> is the secant line between these two points,<br>
whose slope corresponds to the objects <b>average velocity</b> over the interval,<br>
approximately <b>24.5 m/s</b>.<br>
This illustrates how average velocity can be obtained from a secant line on a positiontime graph.


## Tangent Line and Derivative of a Point
### Finding a tangent line to the graph of a function
To find the equation of a tangent line to the graph of a function $y=f(x)$ at a specific point $P(x_0, y_0)$, you need to determine the slope of the curve at that point.  The slope of the curve at $P$ is defined by the derivative of the function at $x_0$, which is given by the limit:

$$m = f'(x_0) = \lim_{h \to 0}\frac{f(x_0 + h) - f(x_0)}{h}$$

Once you have the slope, $m$, and the point $(x_0, y_0)$, you can use the **point-slope formula** to write the equation of the tangent line:

$$y - y_0 = m(x - x_0)$$

This equation represents the line that touches the curve at point $P$ and has the same instantaneous slope as the curve at that single point.

#### Limit of the Difference Quotient

1. **The slope of the graph of \$y=f(x)\$ at \$x=x\_0\$**

   This interpretation highlights the geometric meaning of the derivative. It tells you how steep the curve is at a specific point.

2. **The slope of the tangent line to the curve \$y=f(x)\$ at \$x=x\_0\$**

   This is a more precise geometric interpretation. The tangent line is a linear approximation of the curve at that single point, and its slope is the instantaneous slope of the curve.

3. **The rate of change of \$f(x)\$ with respect to \$x\$ at \$x=x\_0\$**

   This is the physical or applied interpretation. It describes how quickly the output value \$f(x)\$ is changing as the input value \$x\$ changes, at that exact instant.

   For example, if \$f(t)\$ is the position of an object at time \$t\$, then \$f'(t)\$ is its instantaneous velocity.

4. **The derivative \$f'(x\_0)\$ at \$x=x\_0\$**

   This is the formal mathematical name for the limit. It's the central concept in differential calculus, and the other three interpretations are simply different ways of understanding what the derivative represents.

The **derivative of a function** $f(x)$ with respect to the variable $x$ is a new function, denoted as $f'(x)$. Its value at any given $x$ is defined by the following limit, **provided the limit exists**:

<div style="border: 6px solid #0ea5e9; background-color:#f0f9ff; padding: 10px; margin:10px 0;">
$$f'(x) = \lim_{h \to 0}\frac{f(x+h) - f(x)}{h}$$
</div>

This function, <span style="color:emerald;">$f'(x)$</span>, gives the **instantaneous rate of change** of $f(x)$ at every point where it's defined.

---

### <span style="color:#ff6f3c;">Example of Calculating a Derivative from the Definition</span>

Let's find the derivative of the function <span style="color:emerald;">$g(x) = 1 + \sqrt{x}$</span> using the limit definition.


**<span style="color:teal;">Step 1: Set up the difference quotient</span>**

First, we need to find $g(x+h)$.

$$g(x+h) = 1 + \sqrt{x+h}$$


Now, plug this and $g(x)$ into the limit definition:

$$g'(x) = \lim_{h \to 0}\frac{g(x+h) - g(x)}{h} = \lim_{h \to 0}\frac{(1 + \sqrt{x+h}) - (1 + \sqrt{x})}{h}$$




**<span style="color:teal;">Step 2: Simplify the numerator</span>**

The 1-1 = 0 cancel out ,which 1 from $f(x+h) \text{ and } f(x)$:


$$g'(x) = \lim_{h \to 0}\frac{\sqrt{x+h} - \sqrt{x}}{h}$$



**<span style="color:teal;">Step 3: Rationalize using the conjugate</span>**

Multiply top and bottom by <span style="color:emerald;">$(\sqrt{x+h} + \sqrt{x})$</span>:

$$g'(x) = \lim_{h \to 0}\frac{\sqrt{x+h} - \sqrt{x}}{h} \cdot \frac{\sqrt{x+h} + \sqrt{x}}{\sqrt{x+h} + \sqrt{x}}$$


This simplifies:

$$g'(x) = \lim_{h \to 0}\frac{(x+h) - x}{h(\sqrt{x+h} + \sqrt{x})}$$


**<span style="color:teal;">Step 4: Simplify and evaluate the limit</span>**

Cancel $h$:

$$g'(x) = \lim_{h \to 0}\frac{1}{\sqrt{x+h} + \sqrt{x}}$$


Substitute $h=0$:
$$g'(x) = \frac{1}{\sqrt{x} + \sqrt{x}}$$

**<span style="color:emerald;">Step 5: Final Result</span>**

<div style="border-left: 6px solid #059669; background-color:#ecfdf5; padding: 10px; margin:10px 0;">
$$\boxed{g'(x) = \tfrac{1}{2\sqrt{x}}}$$
</div>


<div style="border: 2px solid #facc15; background-color: #fff9e8; padding: 10px; border-radius: 6px;">
  <h3> Notation for Derivatives</h3>
    There are several common notations for the derivative of a function . Each notation is associated with a specific mathematician who helped develop the field of calculus.

1.  **Leibniz's Notation:**
    * This is one of the most common notations, particularly in physics and engineering. It's written as $\frac{dy}{dx}$ or $\frac{d}{dx}f(x)$.
    * It emphasizes that the derivative is the ratio of an infinitesimal change in $y$ to an infinitesimal change in $x$.
    * This notation is excellent for highlighting the independent and dependent variables. For example, if you're taking the derivative with respect to time, you would write $\frac{dy}{dt}$.

2.  **Lagrange's Notation:**
    * This is the notation you used previously: $f'(x)$.
    * It's a very compact and widely used notation, especially in mathematics.
    * The prime symbol ($'$) indicates the first derivative. Higher-order derivatives are denoted with multiple primes, such as $f''(x)$ for the second derivative.

3.  **Newton's Notation:**
    * This notation uses a dot placed above the dependent variable: $\dot{y}$.
    * It is used almost exclusively to denote a derivative with respect to **time**.
    * It is particularly popular in physics and mechanics, where derivatives often represent rates of change over time (e.g., velocity is the time derivative of position).

4.  **Euler's Notation:**
    * This notation uses the operator $D_x$ or simply $D$.
    * For a function $f(x)$, the derivative is written as $D_x f(x)$ or $D_x y$.
    * The notation emphasizes the derivative as an **operator** that acts on a function to produce a new function. For example, $D_x$ "differentiates with respect to $x$."
</div>

## Differentiability on an Interval

A function **$y=f(x)$** is considered **differentiable on an open interval** (finite or infinite) if its derivative exists at every point within that interval.

For a function to be **differentiable on a closed interval** $[a, b]$, two conditions must be met:
1. It must be differentiable on the interior open interval $(a, b)$.
2. The one-sided derivatives at the endpoints must exist. Specifically, the derivative from the right at $a$ and the derivative from the left at $b$ must exist. These are defined by the following limits:
   * **Right-hand derivative at $a$**:
     $f'(a) = \lim_{h \to 0^+} \frac{f(a+h) - f(a)}{h}$
   * **Left-hand derivative at $b$**:
     $f'(b) = \lim_{h \to 0^-} \frac{f(b+h) - f(b)}{h}$

<div style="border: 2px solid #facc15; background-color: #fff9e8; padding: 10px; border-radius: 6px;">
<h3>What Differentiability Means</h3>

For a function to be differentiable at a point, its graph must be both **continuous** and **"smooth"** at that point. This means the curve has no sharp corners, cusps, or vertical tangents.

* **Continuity** ensures there are no breaks or jumps in the graph.
* **Smoothness** ensures that the slope of the curve changes gradually.
<h3>Alternative derivatives of a function also have the alternative version.</h3>
<div style="border: 6px solid #facc15; background-color:#fff9e8; padding: 10px; margin:10px 0; border-radius:6px;">
$$f'(x) = \lim_{z \to x}\frac{f(z) - f(x)}{z-x}$$
</div>
</div>

### <span style="color:#ff6f3c;">Example of Differentiability on an Interval Graph</span>
In this example, we examine the differentiability of several functions over their respective intervals.
The plots illustrate the function itself, its analytical derivative, and various numerical approximations using forward, backward, central, and second-order differences.
We also analyze the errors associated with each method and highlight regions where the function is differentiable.
This visualization demonstrates how numerical differentiation approximates analytical derivatives and how step size influences accuracy.

For this example, we choose the function $f(x) = sin(x^2)$ because it is **smooth** and differentiable over the interval, which allows us to clearly observe the behavior of numerical derivatives.

Other functions, such as $|x|$ or $\sqrt{x}$ have discontinuous or undefined derivatives at certain points, making it harder to visualize the approximation behavior.

```python
from matplotlib.patches import Rectangle
import matplotlib.patches as mpatches

# Set up the plotting style
plt.style.use('seaborn-v0_8')
plt.rcParams['figure.figsize'] = (15, 12)
plt.rcParams['font.size'] = 12

# Define multiple test functions and their analytical derivatives
functions = {
    'sin(x)': {
        'f': lambda x: np.sin(x**2),
        'f_prime': lambda x: 2*x*np.cos(x**2),
        'domain': (-3, 3)
    },
    'x - 3x': {
        'f': lambda x: x**3 - 3*x,
        'f_prime': lambda x: 3*x**2 - 3,
        'domain': (-2.5, 2.5)
    },
    'e^(-x)': {
        'f': lambda x: np.exp(-x**2),
        'f_prime': lambda x: -2*x*np.exp(-x**2),
        'domain': (-3, 3)
    },
    'Absolute value |x|': {
        'f': lambda x: np.abs(x),
        'f_prime': lambda x: np.where(x > 0, 1, np.where(x < 0, -1, np.nan)),  # Undefined at x=0
        'domain': (-2, 2)
    }
}

def numerical_derivatives(f, x, h):
    """Calculate numerical derivatives using different methods"""
    # Forward difference
    forward = (f(x + h) - f(x)) / h
    
    # Backward difference
    backward = (f(x) - f(x - h)) / h
    
    # Central difference
    central = (f(x + h) - f(x - h)) / (2*h)
    
    # Second-order forward difference
    second_order = (-3*f(x) + 4*f(x + h) - f(x + 2*h)) / (2*h)
    
    return forward, backward, central, second_order

def analyze_function(func_name, func_data, h_values=[1e-3, 1e-5, 1e-7]):
    """Analyze differentiability for a given function"""
    f = func_data['f']
    f_prime = func_data['f_prime']
    domain = func_data['domain']
    
    # Create points for evaluation
    x = np.linspace(domain[0], domain[1], 500)
    
    # Create subplot layout
    fig = plt.figure(figsize=(16, 12))
    fig.suptitle(f'Differentiability Analysis: {func_name}', fontsize=16, fontweight='bold')
    
    # Plot 1: Function and its analytical derivative
    ax1 = plt.subplot(2, 3, 1)
    ax1.plot(x, f(x), 'b-', linewidth=3, label=f'f(x) = {func_name}')
    ax1.plot(x, f_prime(x), 'r--', linewidth=2, label="f'(x) analytical")
    ax1.set_title('Function and Analytical Derivative')
    ax1.grid(True, alpha=0.3)
    ax1.legend()
    ax1.set_xlabel('x')
    ax1.set_ylabel('y')
    
    # Plot 2: Numerical derivatives comparison
    ax2 = plt.subplot(2, 3, 2)
    h = h_values[1]  # Use middle h value
    forward, backward, central, second_order = numerical_derivatives(f, x, h)
    
    ax2.plot(x, f_prime(x), 'k-', linewidth=3, label='Analytical', alpha=0.8)
    ax2.plot(x, forward, ':', linewidth=2, label='Forward diff', alpha=0.7)
    ax2.plot(x, backward, '-.', linewidth=2, label='Backward diff', alpha=0.7)
    ax2.plot(x, central, '--', linewidth=2, label='Central diff')
    ax2.plot(x, second_order, '-', linewidth=1, label='2nd order forward', alpha=0.6)
    ax2.set_title(f'Numerical Derivatives (h = {h})')
    ax2.grid(True, alpha=0.3)
    ax2.legend()
    ax2.set_xlabel('x')
    ax2.set_ylabel("f'(x)")
    
    # Plot 3: Error analysis for different h values
    ax3 = plt.subplot(2, 3, 3)
    colors = ['red', 'blue', 'green']
    for i, h in enumerate(h_values):
        forward, backward, central, second_order = numerical_derivatives(f, x, h)
        error_central = np.abs(central - f_prime(x))
        # Remove NaN values for plotting
        valid_mask = ~np.isnan(error_central)
        ax3.semilogy(x[valid_mask], error_central[valid_mask], 
                    color=colors[i], linewidth=2, label=f'h = {h}')
    
    ax3.set_title('Central Difference Error vs Step Size')
    ax3.grid(True, alpha=0.3)
    ax3.legend()
    ax3.set_xlabel('x')
    ax3.set_ylabel('|Error|')
    
    # Plot 4: Step size sensitivity analysis
    ax4 = plt.subplot(2, 3, 4)
    h_range = np.logspace(-8, -1, 50)
    x_test = domain[0] + 0.3 * (domain[1] - domain[0])  # Test point
    errors = []
    
    for h in h_range:
        try:
            _, _, central, _ = numerical_derivatives(f, x_test, h)
            analytical = f_prime(x_test)
            if not np.isnan(analytical):
                error = abs(central - analytical)
                errors.append(error)
            else:
                errors.append(np.nan)
        except:
            errors.append(np.nan)
    
    valid_indices = ~np.isnan(errors)
    ax4.loglog(h_range[valid_indices], np.array(errors)[valid_indices], 'bo-', linewidth=2)
    ax4.set_title(f'Error vs Step Size at x = {x_test:.2f}')
    ax4.grid(True, alpha=0.3)
    ax4.set_xlabel('Step size h')
    ax4.set_ylabel('|Error|')
    
    # Plot 5: Differentiability regions
    ax5 = plt.subplot(2, 3, 5)
    h = h_values[1]
    forward, backward, central, second_order = numerical_derivatives(f, x, h)
    
    # Check where derivatives are well-defined
    analytical_valid = ~np.isnan(f_prime(x))
    numerical_valid = ~np.isnan(central)
    
    ax5.fill_between(x, -1, 1, where=analytical_valid & numerical_valid, 
                    alpha=0.3, color='green', label='Differentiable')
    ax5.fill_between(x, -1, 1, where=~(analytical_valid & numerical_valid), 
                    alpha=0.3, color='red', label='Non-differentiable')
    
    ax5.plot(x, f(x)/max(abs(f(x))), 'b-', linewidth=2, label='f(x) normalized')
    ax5.set_title('Differentiability Regions')
    ax5.grid(True, alpha=0.3)
    ax5.legend()
    ax5.set_xlabel('x')
    ax5.set_ylabel('Normalized values')
    ax5.set_ylim(-1.1, 1.1)
    
    # Plot 6: Error statistics table
    ax6 = plt.subplot(2, 3, 6)
    ax6.axis('off')
    
    # Calculate error statistics
    h = h_values[1]
    forward, backward, central, second_order = numerical_derivatives(f, x, h)
    
    methods = ['Forward', 'Backward', 'Central', '2nd Order Forward']
    derivatives = [forward, backward, central, second_order]
    
    error_stats = []
    for deriv in derivatives:
        error = np.abs(deriv - f_prime(x))
        valid_error = error[~np.isnan(error)]
        if len(valid_error) > 0:
            max_err = np.max(valid_error)
            mean_err = np.mean(valid_error)
            error_stats.append([f'{max_err:.2e}', f'{mean_err:.2e}'])
        else:
            error_stats.append(['N/A', 'N/A'])
    
    # Create table
    table_data = [['Method', 'Max Error', 'Mean Error']]
    for i, method in enumerate(methods):
        table_data.append([method] + error_stats[i])
    
    table = ax6.table(cellText=table_data[1:], colLabels=table_data[0],
                     cellLoc='center', loc='center')
    table.auto_set_font_size(False)
    table.set_fontsize(10)
    table.scale(1.2, 2)
    
    # Style the table
    for i in range(len(table_data)):
        for j in range(len(table_data[0])):
            cell = table[(i, j)]
            if i == 0:  # Header
                cell.set_facecolor('#4472C4')
                cell.set_text_props(weight='bold', color='white')
            else:
                cell.set_facecolor('#F2F2F2')
    
    ax6.set_title('Error Analysis Summary', pad=20)
    
    plt.tight_layout()
    return fig

# Analyze all functions
for func_name, func_data in functions.items():
    fig = analyze_function(func_name, func_data)
    plt.show()
    print(f"\nAnalysis completed for {func_name}")
    print("-" * 50)

# Summary comparison plot
plt.figure(figsize=(12, 8))
x_vals = np.linspace(-2, 2, 300)
h = 1e-5

for i, (name, data) in enumerate(functions.items()):
    f = data['f']
    f_prime = data['f_prime']
    
    # Calculate central difference
    _, _, central, _ = numerical_derivatives(f, x_vals, h)
    error = np.abs(central - f_prime(x_vals))
    
    # Plot only valid errors
    valid_mask = ~np.isnan(error) & (error > 0)
    if np.any(valid_mask):
        plt.semilogy(x_vals[valid_mask], error[valid_mask], 
                    linewidth=2, label=name, alpha=0.8)

plt.title('Numerical Differentiation Error Comparison\n(Central Difference Method)', fontsize=14, fontweight='bold')
plt.xlabel('x', fontsize=12)
plt.ylabel('|Error|', fontsize=12)
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()

print("\n" + "="*60)
print("DIFFERENTIABILITY ANALYSIS COMPLETE")
print("="*60)
print("Key Observations:")
print("1. Central difference typically provides better accuracy than forward/backward")
print("2. Error decreases with smaller step sizes until numerical precision limits")
print("3. Functions with discontinuous derivatives (like |x|) show higher errors")
print("4. Smooth functions (like sin(x), e^(-x)) have lower numerical errors")
print("5. Optimal step size balances truncation and round-off errors")
```

### <span style="color:#ff6f3c;"> Example: One-sided derivatives of $f(x) = \sin(x^2)$ at $x = 1$</span>

**<span style="color:teal;">Step 1: Recall the definitions </span>**

* **Right-hand derivative** at $x = a$:

$$
f'_+(a) = \lim_{h \to 0^+} \frac{f(a+h) - f(a)}{h}
$$

* **Left-hand derivative** at $x = a$:

$$
f'_-(a) = \lim_{h \to 0^-} \frac{f(a+h) - f(a)}{h}
$$

* If $f'_+(a) = f'_-(a)$, then $f$ is **differentiable at $x = a$**.


**<span style="color:teal;"> Step 2: Set up the derivatives at $x = 1$</span>**

* Function: $f(x) = \sin(x^2)$
* Point: $a = 1$

**Right-hand derivative:**

$$
f'_+(1) = \lim_{h \to 0^+} \frac{\sin((1+h)^2) - \sin(1^2)}{h} = \lim_{h \to 0^+} \frac{\sin((1+h)^2) - \sin(1)}{h}
$$

**Left-hand derivative:**

$$
f'_-(1) = \lim_{h \to 0^-} \frac{\sin((1+h)^2) - \sin(1)}{h}
$$


**<span style="color:teal;">Step 3: Expand using the chain rule (optional for exact computation)</span>**

For small $h$, $(1+h)^2 = 1 + 2h + h^2$. Then:

$$
\sin((1+h)^2) = \sin(1 + 2h + h^2)
$$

Using the linear approximation (first-order Taylor expansion):

$$
\sin(1 + 2h + h^2) \approx \sin(1) + \cos(1) \cdot (2h + h^2)
$$


**<span style="color:teal;"> Step 4: Compute the limit</span>**

**Right-hand derivative:**

$$
f'_+(1) \approx \lim_{h \to 0^+} \frac{\sin(1) + \cos(1)(2h + h^2) - \sin(1)}{h} = \lim_{h \to 0^+} \frac{\cos(1)(2h + h^2)}{h}
$$

$$
f'_+(1) = \lim_{h \to 0^+} \cos(1) \cdot \frac{2h + h^2}{h} = \lim_{h \to 0^+} \cos(1) \cdot (2 + h) = 2 \cos(1)
$$

**Left-hand derivative:**

$$
f'_-(1) = \lim_{h \to 0^-} \cos(1) \cdot (2 + h) = 2 \cos(1)
$$


**<span style="color:teal;">Step 5: Conclude</span>**
<div style="border-left: 6px solid #059669; background-color:#ecfdf5; padding: 10px; margin:10px 0;">
<p>$f'_+(1) = f'_-(1) = 2 \cos(1)$</p>
<p>Therefore, $f(x) = \sin(x^2)$ is differentiable at $x = 1$.</p>
</div>



#### When Does a Function Not Have a Derivative at a Point?
A function does not have a derivative at a point if the curve has a sharp change, a break, or a vertical tangent line. For a derivative to exist, the function must be **continuous** and **smooth** at that point.

There are five common reasons why a function might not have a derivative at a specific point along with clear examples:
1. **A Corner**:
     A corner occurs when the graph makes an abrupt change in direction. At this point, the **one-sided derivatives** (the slopes from the left and right) are not equal. Because the slope is different on each side, the overall derivative does not exist.
   $$f(x)=|x|, \text{ which has a corner at } x = 0.$$
2. **A Cusp**: 
    Similar to a corner; a cusp also has a sharp point. However, at a cusp, the slopes of the secant lines approach **infinity** from one side and **negative infinity** from the other.
    $$f(x)=x^{2/3} \text{ at }  x = 0.$$
   The graph looks like a sharp arrowhead pointing down.

3. **A Vertical Tangent Line**:
     A vertical tangent line occurs when the slope of the curve becomes infinitely steep. The slope of the secant lines approaches either positive or negative infinity from both sides of the point.
   $$f(x)=\sqrt{x}, \text{ has a vertical tangent line at } x = 0.$$

4. **A Discontinuity**:
     If a function is not continuous at a point (meaning there's a break or a jump in the graph), it cannot have a derivative there. For a derivative to exist, the function must be defined and connected at that point. A derivative is a limit, and limits can't exist at a break. This includes jump discontinuities and removable discontinuities.

5. **Wild Oscillation**:
     This occurs when a function oscillates so quickly and frequently around a point that the slopes of the secant lines never approach a single, stable value.
   $$f(x) = x \sin\left(\frac{1}{x}\right) \text{ at } x=0. \text{ As you get closer to }  x = 0$$,
the function wiggles back and forth so rapidly that the slope never settles on a single number.

<div style="border: 2px solid #444; border-radius: 8px; padding: 16px; background-color: #fefefe; max-width: 700px; margin: 20px auto; font-family: 'Georgia', serif;"> <h3 style="margin-top: 0; text-align: center;">Proof</h3>
<p style="font-size: 1.05em; line-height: 1.7;"> Suppose $f$ is differentiable at $x = c$. By definition of the derivative, \[ f'(c) = \lim_{x \to c} \frac{f(x) - f(c)}{x - c} \] exists (and is finite). </p> 
<p style="font-size: 1.05em; line-height: 1.7;"> Consider the difference: \[ f(x) - f(c) = \frac{f(x) - f(c)}{x - c} \cdot (x - c), \quad x \neq c. \] Taking the limit as $x \to c$, we get \[ \lim_{x \to c} \big(f(x) - f(c)\big) = \lim_{x \to c} \left[ \frac{f(x) - f(c)}{x - c} \cdot (x - c)\right]. \] </p> 
<p style="font-size: 1.05em; line-height: 1.7;"> Since $\lim_{x \to c} \dfrac{f(x) - f(c)}{x - c} = f'(c)$ (finite), and $\lim_{x \to c} (x - c) = 0$, it follows that \[ \lim_{x \to c} \big(f(x) - f(c)\big) = f'(c)\cdot 0 = 0. \] </p> 
<p style="font-size: 1.05em; line-height: 1.7;"> Therefore, \[ \lim_{x \to c} f(x) = f(c). \] This shows $f$ is continuous at $x = c$. </p> 
<p style="font-weight: bold; text-align: center; margin-top: 1em;"> </p> 
</div>
$$
\boxed{\;\;\; \therefore f \text{ differentiable at } c \implies f \text{ continuous at } c \;\;\;}
$$

### <span style="color:#ff6f3c;">Example: Given $f(x) = (x-1)^2 + 1$,</span>
<span>Calculate  the derivative using the **definition of derivative**. <span>

**<span style="color:teal;">Step 1: Recall the definition</span>**

* Derivative at $x = a$:

$$
f'(a) = \lim_{h \to 0} \frac{f(a+h) - f(a)}{h}
$$


**<span style="color:teal;">Step 2: Set up the derivative at $x$</span>**

* Function: $f(x) = (x-1)^2 + 1$
* Point: $a = x$ (general $x$)

$$
f'(x) = \lim_{h \to 0} \frac{f(x+h) - f(x)}{h}
= \lim_{h \to 0} \frac{(x+h-1)^2 + 1 - [(x-1)^2 + 1]}{h}
$$


**<span style="color:teal;">Step 3: Expand the square</span>**

$$
(x+h-1)^2 = (x-1)^2 + 2h(x-1) + h^2
$$

So the numerator becomes:

$$
f(x+h) - f(x) = 2h(x-1) + h^2
$$


**<span style="color:teal;">Step 4: Divide by $h$ and compute the limit</span>**

$$
\frac{f(x+h)-f(x)}{h} = \frac{2h(x-1)+h^2}{h} = 2(x-1) + h
$$

$$
f'(x) = \lim_{h \to 0} [2(x-1)+h] = 2(x-1) = 2x - 2
$$


**<span style="color:teal;">Step 5: Conclude</span>**

<div style="border-left: 6px solid #059669; background-color:#ecfdf5; padding: 10px; margin:10px 0;">
<p>\(f'(x) =  2x - 2\)</p>
<p>Therefore, \(f(x) = (x-1)^2 + 1\) is differentiable for all \(x\).</p>
</div>
