# Rage of Change and Tangent Line
The rate of change is a fundamental concept in physics, often seen in everyday life as the rate of change of speed, which is **acceleration**.

## Average Speed

When we measure speed over a time interval, we compare the position of an object at one moment in time to its position at another moment. For example, if a falling object begins to drop at **14:00:00** and we observe its position a few seconds later (**14:00:10**), the **average speed** is calculated by dividing the change in distance by the change in time.

We can express this as a function:

$$
\text{Average Speed}(t_1, t_2) = \frac{f(t_2) - f(t_1)}{t_2 - t_1},
$$

where $f(t)$ represents the position of the object at time $t$.

If we assume the motion starts at $t_1 = x$ and ends at $t_2 = x+h$, then the formula becomes

$$
\text{Average Speed}(x, x+h) = \frac{f(x+h) - f(x)}{(x+h) - x}.
$$

Simplifying the denominator gives

$$
\text{Average Speed}(x, x+h) = \frac{f(x+h) - f(x)}{h}.
$$

This form is the **difference quotient**, and as $h \to 0$, it leads directly to the definition of the **instantaneous speed** (derivative).

```python
# Define position function for a falling object under gravity (ignoring air resistance)
g = 9.8
f = lambda t: 0.5 * g * t**2

# Time range
t = np.linspace(1.75, 3.25, 100)
y = f(t)

# Pick a point x and interval h for difference quotient
x = 2   # seconds
h = 1   # seconds

# Average speed between x and x+h
avg_speed = (f(x+h) - f(x)) / h

# Values for secant line
secant_t = np.array([x, x+h])
secant_y = np.array([f(x), f(x+h)])

# Plot
plt.figure(figsize=(8,6))
plt.plot(t, y, label=r"$f(t) = \frac{1}{2} g t^2$", color="dodgerblue")
plt.scatter([x, x+h], [f(x), f(x+h)], color="red", zorder=5, label="Points (x, f(x)) & (x+h, f(x+h))")
plt.plot(secant_t, secant_y, "--", color="palegreen", label=f"Secant line (avg speed ≈ {avg_speed:.2f} m/s)")

# Mark h interval on x-axis
plt.hlines(f(x), x, x+h, colors="gray", linestyles="dotted")
plt.vlines(x, 0, f(x), colors="gray", linestyles="dotted")
plt.vlines(x+h, 0, f(x+h), colors="gray", linestyles="dotted")

plt.title("Falling Object and Average Speed (Secant Line)")
plt.xlabel("Time (s)")
plt.ylabel("Position f(t) (m)")
plt.legend()
plt.grid(True)
plt.show()
```

## Tangent line and Derivative of Point
### Finding a tangent line to the graph of a function
To find the equation of a tangent line to the graph of a function $y=f(x)$ at a specific point $P(x_0, y_0)$, you need to determine the slope of the curve at that point.  The slope of the curve at $P$ is defined by the derivative of the function at $x_0$, which is given by the limit:

$$m = f'(x_0) = \lim_{h \to 0}\frac{f(x_0 + h) - f(x_0)}{h}$$

Once you have the slope, $m$, and the point $(x_0, y_0)$, you can use the **point-slope formula** to write the equation of the tangent line:

$$y - y_0 = m(x - x_0)$$

This equation represents the line that just touches the curve at point $P$ and has the same instantaneous slope as the curve at that single point.

#### Limit of the Difference Quotient

1. **The slope of the graph of \$y=f(x)\$ at \$x=x\_0\$**

   This interpretation highlights the geometric meaning of the derivative. It tells you how steep the curve is at a specific point.

2. **The slope of the tangent line to the curve \$y=f(x)\$ at \$x=x\_0\$**

   This is a more precise geometric interpretation. The tangent line is a linear approximation of the curve at that single point, and its slope is the instantaneous slope of the curve.

3. **The rate of change of \$f(x)\$ with respect to \$x\$ at \$x=x\_0\$**

   This is the physical or applied interpretation. It describes how quickly the output value \$f(x)\$ is changing as the input value \$x\$ changes, at that exact instant.

   For example, if \$f(t)\$ is the position of an object at time \$t\$, then \$f'(t)\$ is its instantaneous velocity.

4. **The derivative \$f'(x\_0)\$ at \$x=x\_0\$**

   This is the formal mathematical name for the limit. It's the central concept in differential calculus, and the other three interpretations are simply different ways of understanding what the derivative represents.

The **derivative of a function** $f(x)$ with respect to the variable $x$ is a new function, denoted as $f'(x)$. Its value at any given $x$ is defined by the following limit, **provided the limit exists**:

<div style="border: 6px solid #0ea5e9; background-color:#f0f9ff; padding: 10px; margin:10px 0;">
$$f'(x) = \lim_{h \to 0}\frac{f(x+h) - f(x)}{h}$$
</div>

This function, <span style="color:emerald;">$f'(x)$</span>, gives the **instantaneous rate of change** of $f(x)$ at every point where it's defined.

---

### <span style="color:#ff6f3c;">Example of Calculating a Derivative from the Definition</span>

Let's find the derivative of the function <span style="color:emerald;">$g(x) = 1 + \sqrt{x}$</span> using the limit definition.


**<span style="color:teal;">Step 1: Set up the difference quotient</span>**

First, we need to find $g(x+h)$.

$$g(x+h) = 1 + \sqrt{x+h}$$


Now, plug this and $g(x)$ into the limit definition:

$$g'(x) = \lim_{h \to 0}\frac{g(x+h) - g(x)}{h} = \lim_{h \to 0}\frac{(1 + \sqrt{x+h}) - (1 + \sqrt{x})}{h}$$




**<span style="color:teal;">Step 2: Simplify the numerator</span>**

The 1s cancel out.

$$g'(x) = \lim_{h \to 0}\frac{\sqrt{x+h} - \sqrt{x}}{h}$$



**<span style="color:teal;">Step 3: Rationalize using the conjugate</span>**

Multiply top and bottom by <span style="color:emerald;">$(\sqrt{x+h} + \sqrt{x})$</span>:

$$g'(x) = \lim_{h \to 0}\frac{\sqrt{x+h} - \sqrt{x}}{h} \cdot \frac{\sqrt{x+h} + \sqrt{x}}{\sqrt{x+h} + \sqrt{x}}$$


This simplifies:

$$g'(x) = \lim_{h \to 0}\frac{(x+h) - x}{h(\sqrt{x+h} + \sqrt{x})}$$


**<span style="color:teal;">Step 4: Simplify and evaluate the limit</span>**

Cancel $h$:

$$g'(x) = \lim_{h \to 0}\frac{1}{\sqrt{x+h} + \sqrt{x}}$$


Substitute $h=0$:
$$g'(x) = \frac{1}{\sqrt{x} + \sqrt{x}}$$

**<span style="color:emerald;">Step 5: Final Result</span>**

<div style="border-left: 6px solid #059669; background-color:#ecfdf5; padding: 10px; margin:10px 0;">
$$\boxed{g'(x) = \tfrac{1}{2\sqrt{x}}}$$
</div>


<div style="border: 2px solid #facc15; background-color: #fff9e8; padding: 10px; border-radius: 6px;">
  <h3>📕 Notation for Derivatives</h3>
    there are several common notations for the derivative of a function $y=f(x)$. Each notation is associated with a specific mathematician who helped develop the field of calculus.

1.  **Leibniz's Notation:**
    * This is one of the most common notations, particularly in physics and engineering. It's written as $\frac{dy}{dx}$ or $\frac{d}{dx}f(x)$.
    * It emphasizes that the derivative is the ratio of an infinitesimal change in $y$ to an infinitesimal change in $x$.
    * This notation is excellent for highlighting the independent and dependent variables. For example, if you're taking the derivative with respect to time, you would write $\frac{dy}{dt}$.

2.  **Lagrange's Notation:**
    * This is the notation you used previously: $f'(x)$.
    * It's a very compact and widely used notation, especially in mathematics.
    * The prime symbol ($'$) indicates the first derivative. Higher-order derivatives are denoted with multiple primes, such as $f''(x)$ for the second derivative.

3.  **Newton's Notation:**
    * This notation uses a dot placed above the dependent variable: $\dot{y}$.
    * It is used almost exclusively to denote a derivative with respect to **time**.
    * It is particularly popular in physics and mechanics, where derivatives often represent rates of change over time (e.g., velocity is the time derivative of position).

4.  **Euler's Notation:**
    * This notation uses the operator $D_x$ or simply $D$.
    * For a function $f(x)$, the derivative is written as $D_x f(x)$ or $D_x y$.
    * The notation emphasizes the derivative as an **operator** that acts on a function to produce a new function. For example, $D_x$ "differentiates with respect to $x$."
</div>

## Differentiability on an Interval

A function **$y=f(x)$** is considered **differentiable on an open interval** (finite or infinite) if its derivative exists at every point within that interval.

For a function to be **differentiable on a closed interval** $[a, b]$, two conditions must be met:
1. It must be differentiable on the interior open interval $(a, b)$.
2. The one-sided derivatives at the endpoints must exist. Specifically, the derivative from the right at $a$ and the derivative from the left at $b$ must exist. These are defined by the following limits:
   * **Right-hand derivative at $a$**:
     $f'(a) = \lim_{h \to 0^+} \frac{f(a+h) - f(a)}{h}$
   * **Left-hand derivative at $b$**:
     $f'(b) = \lim_{h \to 0^-} \frac{f(b+h) - f(b)}{h}$

<div style="border: 2px solid #facc15; background-color: #fff9e8; padding: 10px; border-radius: 6px;">
<h3>📕What Differentiability Means</h3>

For a function to be differentiable at a point, its graph must be both **continuous** and **"smooth"** at that point. This means the curve has no sharp corners, cusps, or vertical tangents. .

* **Continuity** ensures there are no breaks or jumps in the graph.
* **Smoothness** ensures that the slope of the curve changes gradually.
</div>

```python
# Define the function and its analytical derivative
f = lambda x: np.sin(x**2)
f_prime = lambda x: 2*x*np.cos(x**2)

# Points to evaluate
x = np.linspace(-3, 3, 400)
h = 1e-5

# One-sided and central derivatives
left_deriv = (f(x) - f(x - h)) / h
right_deriv = (f(x + h) - f(x)) / h
central_deriv = (f(x + h) - f(x - h)) / (2*h)

# Plot the function
plt.figure(figsize=(10, 6))
plt.plot(x, f(x), label='f(x) = sin(x²)', color='navy')
plt.scatter(0, f(0), color='red', zorder=10)
plt.title('Function and Derivatives')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

# Plot the derivatives
plt.plot(x, f_prime(x), label='Analytical', color='black', linewidth=2)
plt.plot(x, left_deriv, '--', label='Left-hand')
plt.plot(x, right_deriv, '-.', label='Right-hand')
plt.plot(x, central_deriv, ':', label='Central')

plt.legend()
plt.show()

# Simple error analysis
errors_left = np.abs(left_deriv - f_prime(x))
errors_right = np.abs(right_deriv - f_prime(x))
errors_central = np.abs(central_deriv - f_prime(x))

print(f"Max error (Left): {errors_left.max():.2e}")
print(f"Max error (Right): {errors_right.max():.2e}")
print(f"Max error (Central): {errors_central.max():.2e}")
```

### <span style="color:#ff6f3c;"> Example: One-sided derivatives of $f(x) = \sin(x^2)$ at $x = 1$</span>

**<span style="color:teal;">Step 1: Recall the definitions </span>**

* **Right-hand derivative** at $x = a$:

$$
f'_+(a) = \lim_{h \to 0^+} \frac{f(a+h) - f(a)}{h}
$$

* **Left-hand derivative** at $x = a$:

$$
f'_-(a) = \lim_{h \to 0^-} \frac{f(a+h) - f(a)}{h}
$$

* If $f'_+(a) = f'_-(a)$, then $f$ is **differentiable at $x = a$**.


**<span style="color:teal;"> Step 2: Set up the derivatives at $x = 1$</span>**

* Function: $f(x) = \sin(x^2)$
* Point: $a = 1$

**Right-hand derivative:**

$$
f'_+(1) = \lim_{h \to 0^+} \frac{\sin((1+h)^2) - \sin(1^2)}{h} = \lim_{h \to 0^+} \frac{\sin((1+h)^2) - \sin(1)}{h}
$$

**Left-hand derivative:**

$$
f'_-(1) = \lim_{h \to 0^-} \frac{\sin((1+h)^2) - \sin(1)}{h}
$$


**<span style="color:teal;">Step 3: Expand using the chain rule (optional for exact computation)</span>**

For small $h$, $(1+h)^2 = 1 + 2h + h^2$. Then:

$$
\sin((1+h)^2) = \sin(1 + 2h + h^2)
$$

Using the linear approximation (first-order Taylor expansion):

$$
\sin(1 + 2h + h^2) \approx \sin(1) + \cos(1) \cdot (2h + h^2)
$$


**<span style="color:teal;"> Step 4: Compute the limit</span>**

**Right-hand derivative:**

$$
f'_+(1) \approx \lim_{h \to 0^+} \frac{\sin(1) + \cos(1)(2h + h^2) - \sin(1)}{h} = \lim_{h \to 0^+} \frac{\cos(1)(2h + h^2)}{h}
$$

$$
f'_+(1) = \lim_{h \to 0^+} \cos(1) \cdot \frac{2h + h^2}{h} = \lim_{h \to 0^+} \cos(1) \cdot (2 + h) = 2 \cos(1)
$$

**Left-hand derivative:**

$$
f'_-(1) = \lim_{h \to 0^-} \cos(1) \cdot (2 + h) = 2 \cos(1)
$$


**<span style="color:teal;">Step 5: Conclude</span>**
<div style="border-left: 6px solid #059669; background-color:#ecfdf5; padding: 10px; margin:10px 0;">
<p>$f'_+(1) = f'_-(1) = 2 \cos(1)$</p>
<p>Therefore, $f(x) = \sin(x^2)$ is differentiable at $x = 1$.</p>
</div>



#### When Does a Function Not Have a Derivative at a Point?
A function does not have a derivative at a point if the curve has a sharp change, a break, or a vertical tangent line. For a derivative to exist, the function must be **continuous** and **smooth** at that point.

There are five common reasons why a function might not have a derivative at a specific point along with clear examples:
1. **A Corner**
    A corner occurs when the graph makes an abrupt change in direction. At this point, the **one-sided derivatives** (the slopes from the left and right) are not equal. Because the slope is different on each side, the overall derivative does not exist.
   $$f(x)=|x|, \text{ which has a corner at } x = 0.$$
3. **A Cusp** 
   Similar to a corner, a cusp also has a sharp point. However, at a cusp, the slopes of the secant lines approach **infinity** from one side and **negative infinity** from the other.
    $$f(x)=x^{2/3} \text{ at }  x = 0.$$
   The graph looks like a sharp arrowhead pointing down.

5. **A Vertical Tangent Line**
    A vertical tangent line occurs when the slope of the curve becomes infinitely steep. The slope of the secant lines approaches either positive or negative infinity from both sides of the point.
   $$f(x)=\sqrt{x}, \text{ has a vertical tangent line at } x = 0.$$

7. **A Discontinuity**
    If a function is not continuous at a point (meaning there's a break or a jump in the graph), it cannot have a derivative there. For a derivative to exist, the function must be defined and connected at that point. A derivative is a limit, and limits can't exist at a break. This includes jump discontinuities and removable discontinuities.

8. **Wild Oscillation**
    This occurs when a function oscillates so quickly and frequently around a point that the slopes of the secant lines never approach a single, stable value.
   $$f(x) = x \sin\left(\frac{1}{x}\right) \text{ at } x=0. \text{ As you get closer to }  x = 0$$,
the function wiggles back and forth so rapidly that the slope never settles on a single number.
