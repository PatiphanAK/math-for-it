# Functions, Graphs, and Models

## Function and Their Graph
#### What is function???
In general, a function is a relationship between inputs and outputs where each input is associated with exactly one output
##### Example #####
![Function](../images/func1.png)
#### But in Math!! ####
A function is a special relationship between
"
- a set of inputs, called the **Domain**, 
- a set of outputs, called the **Range**
".

### Advanced Perspective ##
A function can also be viewed as an **infinite-dimensional vector**.  
We will explore this idea in **Functional Analysis**, where the basic proof can be established using the *Fourier Transform*.

#### Example ####
**Domain** (Inputs): Let's say our domain is the set of numbers 
$$ A = \{1, 2, 3\} $$
These are the inputs we are allowed to use.
Then the Function we have is.
$$
f(x) = 2(x)+1
$$
Applying the Function:
$$
\begin{aligned}
f(1)=2(1)+1=3  \\
f(2)=2(2)+1=5  \\
f(3)=2(3)+1=7  \\
\end{aligned}
$$
**Range** (Outputs): The set of all outputs we got is 
$$ B = \{3,5,7\} $$
This is the range of the function for the given domain.

```python
import numpy as np
import matplotlib.pyplot as plt
```

### example graph of function
$$y = 2{x}+1$$

```python
## This is example how we write function in python
def f(x):
    return 2*x+1

x = np.linspace(-5, 5, 100)
y = f(x)
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='y = 2x+1', color='blue', linestyle='-')
plt.title('Graph of y = 2x + 1')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.grid(True, linestyle=':', alpha=0.6)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.show()
```

### Trigonometric Function
Angles are measured in degrees or radians
#### Trigonometric Identities
$$
\cos^2\theta + \sin^2\theta = 1
$$
$$
1 + \tan^2\theta = \sec^2\theta
$$
$$
1 + \cot^2\theta = \csc^2\theta
$$
#### Addition Formulas
$$
\begin{aligned}
\cos(A+B) = \cos(A)\cos(B) - \sin(A)\sin(B) \\
\sin(A+B) = \sin(A)\cos(B) + \cos(A)\sin(B)
\end{aligned}
$$
#### Double-Angle Formulas
$$
\begin{aligned}
\cos2\theta &= \cos^2\theta - \sin^2\theta \\
\sin2\theta &= 2\sin\theta\cos\theta
\end{aligned}
$$
#### Half-Angle Formulas
$$
\begin{aligned}
\cos^2\theta = \frac{1+\cos2\theta}2 \\
\sin^2\theta = \frac{1-\cos2\theta}2
\end{aligned}
$$
#### The Law of Cosines
$$
c^2 = a^2 + b^2 - 2ab\cos\theta
$$
#### Two Special Inequalities
$$
\begin{aligned}
-|\theta| \le \sin\theta \le |\theta| \\
-|\theta| \le 1-\cos\theta \le |\theta| 
\end{aligned}
$$

```python
theta = np.linspace(0, 2 * np.pi, 100)
x = np.cos(theta)
y = np.sin(theta)
plt.figure(figsize=(8, 8))
plt.plot(x, y, color='blue', linewidth=2)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.arrow(0, 0, 1.2, 0, head_width=0.05, head_length=0.08, fc='black', ec='black')
plt.arrow(0, 0, 0, 1.2, head_width=0.05, head_length=0.08, fc='black', ec='black')
plt.xlabel('X-axis', fontsize=12)
plt.ylabel('Y-axis', fontsize=12)
plt.title('Unit Circle', fontsize=16)
plt.axis('equal')
plt.grid(True, linestyle=':', alpha=0.6)
plt.show()
```

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 400)
y1 = np.sin(x)+5
plt.figure(figsize=(10, 6))
plt.plot(x, y1, label=r'$f(x) = \sin(x) + 5$', color='blue', linewidth=2)
plt.axhline(5, color='red', linestyle='--', linewidth=1, label='y = 5')
plt.xlabel('x', fontsize=14)
plt.ylabel('f(x)', fontsize=14)
plt.title(r'Graph of $f(x) = \sin(x) + 5$', fontsize=16)
plt.legend(fontsize=12)
plt.grid(True, linestyle=':', alpha=0.6)
plt.ylim(3, 7)
plt.show()
```

```python
y2 = np.cos(x)+5
plt.figure(figsize=(10, 6))
plt.plot(x, y1, label=r'$f(x) = \sin(x) + 5$', color='blue', linewidth=2)
plt.plot(x, y2, label=r'$f(x) = \cos(x) + 5$', color='green', linewidth=2)
plt.axhline(5, color='red', linestyle='--', linewidth=1, label='y = 5')
plt.xlabel('x', fontsize=14)
plt.ylabel('f(x)', fontsize=14)
plt.title(r'Graph of $f(x) = \sin(x) + 5$ and $\cos(x) + 5$', fontsize=16)
plt.legend(fontsize=12)
plt.grid(True, linestyle=':', alpha=0.6)
plt.ylim(3, 7)
plt.show()
```
