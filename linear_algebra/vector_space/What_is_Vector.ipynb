## Vector in $\mathbb{R}^n$
A **vector** in $\mathbb{R}^n$ is an ordered list of n real numbers:

$$\mathbf{v} = (v_1, v_2, \dots, v_n)$$

Order matters: (2, 3) and (3, 2) are different vectors.

**Example:** Patiphan's house is at (2, 3) ‚Äî 2 units east, 3 units north.

### The Mathematical View

Formally, a vector in $\mathbb{R}^n$ is an element of the set:

$$\mathbb{R}^n = \{(x_1, x_2, \dots, x_n) : x_i \in \mathbb{R}\}$$

Two vectors are **equal** if and only if every corresponding component is equal:

$$(a_1, \dots, a_n) = (b_1, \dots, b_n) \iff a_i = b_i \text{ for all } i$$

**Notation:** $\mathbb{R}$ denotes the set of real numbers,
and $\mathbb{R}^n$ denotes all ordered lists of $n$ real numbers.

## Difference Between Dimension in Mathematics and NumPy

In NumPy, the function `ndim` returns the number of axes (also called tensor rank) of an array.

Example:

```python
import numpy as np

v = np.array([1, 2, 3])
print(v.ndim)
````

Output:

```
1
```

This means the array has **one axis**.

However, in mathematics (Linear Algebra), the word "dimension" means something different.

The vector

$$
v = (1,2,3)
$$

is an element of

$$
\mathbb{R}^3
$$

The dimension of the vector space $\mathbb{R}^3$ is:

$$
3
$$

because it has 3 basis vectors.

---

### Key Difference

| Concept | NumPy | PyTorch | Mathematics |
|---|---|---|---|
| Meaning of "dimension" | Number of axes | Number of axes | Number of basis vectors |
| For `[1,2,3]` | `ndim = 1` | `dim() = 1` | Space dimension = 3 |
| Shape | `(3,)` | `torch.Size([3])` | Vector in $\mathbb{R}^3$ |

---

### Important Insight

NumPy dimension describes the **structure of the data container**.

Mathematical dimension describes the **dimension of the vector space**.

These are not the same concept.

# üìç Example Problem in 2D: Coordinates and Distance

---

<div style="
background-color:#f4f8ff;
border-left:6px solid #4a6fa5;
padding:12px;
border-radius:6px;
">

<b>What is a Coordinate?</b><br><br>

A coordinate system allows us to describe the position of a point in space using numbers. In two-dimensional space ($\mathbb{R}^2$), a point is written as:

$$
(x, y)
$$

where $x$ represents the **horizontal position** and $y$ represents the **vertical position**.

**Example:** The geographic coordinate system used on Earth is a real-world example of coordinates. Every location on Earth is described by two numbers:
- **Latitude** ‚Äî the north‚Äìsouth position (e.g., 51.5¬∞ N)
- **Longitude** ‚Äî the east‚Äìwest position (e.g., 0.0¬∞, the Greenwich meridian)

So the location of Greenwich, London is $(0¬∞, 51.5¬∞)$ ‚Äî a coordinate pair, just like $(x, y)$.

</div>

---

<div style="
background-color:#fffaf0;
border-left:6px solid #c9a227;
padding:12px;
border-radius:6px;
">

<b>Problem.</b><br><br>

Suppose:

- Patiphan's house is located at  
  $$
  (x_1, y_1)
  $$

- Worrawat's house is located at  
  $$
  (x_2, y_2)
  $$

Find the distance between the two houses.

</div>

---

<div style="
background-color:#f1f8f4;
border-left:6px solid #4f8a5b;
padding:12px;
border-radius:6px;
">

<b>Solution.</b><br><br>

We interpret each location as a vector in $\mathbb{R}^2$:

$$
\mathbf{p} = (x_1, y_1)
$$

$$
\mathbf{q} = (x_2, y_2)
$$

The vector pointing from Patiphan's house to Worrawat's house is:

$$
\mathbf{q} - \mathbf{p}
=
(x_2 - x_1,\; y_2 - y_1)
$$

The distance is the norm (length) of this vector:

$$
\color{#4a6fa5}{
d =
\sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2}
}
$$

</div>

---

<div style="
background-color:#f8f5ff;
border-left:6px solid #7b6fc2;
padding:12px;
border-radius:6px;
">

<b>Interpretation.</b><br><br>

The distance formula comes from the Euclidean norm in $\mathbb{R}^2$:

$$
\| (a, b) \|
=
\sqrt{a^2 + b^2}
$$

So geometry is naturally expressed using vectors.

</div>

`````python
import math

patiphan_house = (2,3)
worrawat_house = (7,9)

def euclidean_distance_basic(p, q):
    total = 0
    for i in range(len(p)):
        total += (q[i] - p[i])**2
    return math.sqrt(total)

print(f"Distance between houses (No Library) {euclidean_distance_basic(patiphan_house, worrawat_house)}")
`````

`````python
import numpy as np
import matplotlib.pyplot as plt

patiphan_house = np.array([2, 3])
worrawat_house = np.array([7, 9])

def euclidean_distance_numpy(p, q):
    v = q - p
    return np.sqrt(np.sum(v**2))
    
dist = euclidean_distance_numpy(patiphan_house, worrawat_house)
print(f"Distance between houses (NumPy): {dist:.4f}")
`````

`````python
# --- Visualization ---
fig, ax = plt.subplots(figsize=(7, 6))

ax.scatter(*patiphan_house, color='royalblue', s=120, zorder=5)
ax.scatter(*worrawat_house, color='tomato',    s=120, zorder=5)

ax.annotate("Patiphan\n(2, 3)", patiphan_house,
            textcoords="offset points", xytext=(-15, -25), fontsize=10, color='royalblue')
ax.annotate("Worrawat\n(7, 9)", worrawat_house,
            textcoords="offset points", xytext=(8, 5),    fontsize=10, color='tomato')

ax.plot(*zip(patiphan_house, worrawat_house), 'k--', linewidth=1.5, label=f'Distance = {dist:.4f}')

ax.plot([patiphan_house[0], worrawat_house[0]], [patiphan_house[1], patiphan_house[1]],
        color='gray', linestyle=':', linewidth=1)
ax.plot([worrawat_house[0], worrawat_house[0]], [patiphan_house[1], worrawat_house[1]],
        color='gray', linestyle=':', linewidth=1)

ax.text(4.5, 2.5, f'Œîx = {worrawat_house[0]-patiphan_house[0]}', ha='center', fontsize=9, color='gray')
ax.text(7.4, 6.0, f'Œîy = {worrawat_house[1]-patiphan_house[1]}', ha='center', fontsize=9, color='gray')

ax.set_xlim(0, 10); ax.set_ylim(0, 12)
ax.set_xlabel('X'); ax.set_ylabel('Y')
ax.set_title('Euclidean Distance Between Two Houses', fontsize=13, fontweight='bold')
ax.legend(fontsize=10)
ax.grid(True, linestyle='--', alpha=0.4)
ax.set_aspect('equal')

plt.tight_layout()
plt.show()
`````

## üìç Example Problem in ‚ÑÇ¬π: Distance with Complex Numbers

So far we measured distance in ‚Ñù¬≤ using real coordinates.
But vectors can also have **complex** components.

In ‚ÑÇ, a number looks like:
$$z = a + bi \quad a, b \in \mathbb{R},\; i = \sqrt{-1}$$

The **modulus** (length) of z is:
$$|z| = \sqrt{a^2 + b^2} = \sqrt{z \cdot \bar{z}}$$

where $\bar{z} = a - bi$ is the **complex conjugate**.

---

### Why not just use (a¬≤ + b¬≤)?

For real numbers: $\|v\|^2 = \sum v_i^2$

For complex numbers, this breaks ‚Äî we need:
$$\|v\|^2 = \sum |v_i|^2 = \sum v_i \overline{v_i}$$

This is called the **Hermitian norm** (or Euclidean norm over ‚ÑÇ).

---

### Problem

Patiphan's position in signal space: $\mathbf{p} = (2+i,\; 3-2i)$

Worrawat's position in signal space: $\mathbf{w} = (5+3i,\; 1+i)$

Find the distance $\|\mathbf{w} - \mathbf{p}\|$.

### Solution

$$\mathbf{w} - \mathbf{p} = (3+2i,\; -2+3i)$$

$$\|\mathbf{w} - \mathbf{p}\|^2 = |3+2i|^2 + |-2+3i|^2$$

$$= (9+4) + (4+9) = 13 + 13 = 26$$

$$\|\mathbf{w} - \mathbf{p}\| = \sqrt{26} \approx 5.099$$

`````python
p = np.array([2+1j, 3-2j])
w = np.array([5+3j, 1+1j])

v = w - p
dist = np.sqrt(np.sum(np.abs(v)**2))

print(f"v = w - p = {v}")
print(f"|v|¬≤ = {np.abs(v)**2}")
print(f"distance = {dist:.4f}")
`````

### Key Insight

| | Real ‚Ñù‚Åø | Complex ‚ÑÇ‚Åø |
|---|---|---|
| Norm¬≤ | $\sum v_i^2$ | $\sum \|v_i\|^2 = \sum v_i\bar{v_i}$ |
| NumPy | `np.sum(v**2)` | `np.sum(np.abs(v)**2)` |
| Why different? | $v_i^2 \geq 0$ always | $v_i^2$ can be negative in ‚ÑÇ |

<div style="
background-color:#fffaf0;
border-left:6px solid #c9a227;
padding:16px;
border-radius:6px;
">

<b>What is an Argand Diagram?</b><br><br>

Just like we plot $(x, y)$ on a 2D graph, we can plot a complex number $z = a + bi$ on a special graph called an <b>Argand diagram</b>:

- The <b>horizontal axis</b> = real part ($a$)
- The <b>vertical axis</b> = imaginary part ($b$)

So $z = 3 + 2i$ is plotted at the point $(3, 2)$.

<br>

This means every complex number <b>is secretly a point in 2D space</b> ‚Äî
just like the coordinates of Patiphan's house!

</div>

---

<div style="
background-color:#f1f8f4;
border-left:6px solid #4f8a5b;
padding:16px;
border-radius:6px;
">

<b>How do we measure the length of a complex number?</b><br><br>

The <b>modulus</b> (length) of $z = a + bi$ is the distance from the origin to the point $(a, b)$:

$$|z| = \sqrt{a^2 + b^2}$$

This is exactly the same as Euclidean distance we already know!<br><br>

<b>Example:</b> $|3 + 2i| = \sqrt{3^2 + 2^2} = \sqrt{13} \approx 3.606$

<br>

And the <b>complex conjugate</b> of $z = a + bi$ is:

$$\bar{z} = a - bi \quad \text{(just flip the sign of imaginary part)}$$

A useful trick: $\; z \cdot \bar{z} = a^2 + b^2 = |z|^2$

</div>

---

<div style="
background-color:#f8f5ff;
border-left:6px solid #7b6fc2;
padding:16px;
border-radius:6px;
">

<b>Back to our problem</b><br><br>

Now instead of houses on a map,
imagine Patiphan and Worrawat are at positions in a <b>complex signal space</b>
where each position has 2 complex components:

$$\mathbf{p} = (2+i,\; 3-2i) \in \mathbb{C}^2$$
$$\mathbf{w} = (5+3i,\; 1+i) \in \mathbb{C}^2$$

To find the distance, we use the <b>Hermitian norm</b>:

$$\|\mathbf{w} - \mathbf{p}\| = \sqrt{|v_1|^2 + |v_2|^2}$$

where $v_1, v_2$ are the components of $\mathbf{w} - \mathbf{p}$.<br><br>

This is the same idea as 2D distance ‚Äî just extended to complex numbers!

</div>

`````python
# --- Data ---
p = np.array([2+1j, 3-2j])
w = np.array([5+3j, 1+1j])
v = w - p  # [3+2j, -2+3j]
dist = np.sqrt(np.sum(np.abs(v)**2))

# --- Figure: 1 row, 3 cols ---
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
fig.suptitle('Distance in ‚ÑÇ‚Åø ‚Äî Hermitian Norm', fontsize=14, fontweight='bold')

# ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
# Plot 1: Component 1  (p[0], w[0])
# ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
ax = axes[0]
ax.set_title('Component 1', fontsize=11)

pts = {'p‚ÇÅ = 2+i': p[0], 'w‚ÇÅ = 5+3i': w[0]}
colors = {'p‚ÇÅ = 2+i': 'royalblue', 'w‚ÇÅ = 5+3i': 'tomato'}

for label, z in pts.items():
    ax.scatter(z.real, z.imag, s=120, color=colors[label], zorder=5)
    ax.annotate(label, (z.real, z.imag),
                textcoords='offset points', xytext=(8, 5), fontsize=9, color=colors[label])

# vector w-p on component 1
ax.annotate('', xy=(w[0].real, w[0].imag), xytext=(p[0].real, p[0].imag),
            arrowprops=dict(arrowstyle='->', color='black', lw=1.5))

# Re/Im dashed components of v[0]
ax.plot([p[0].real, w[0].real], [p[0].imag, p[0].imag],
        color='gray', linestyle=':', linewidth=1)
ax.plot([w[0].real, w[0].real], [p[0].imag, w[0].imag],
        color='gray', linestyle=':', linewidth=1)

ax.text((p[0].real + w[0].real)/2, p[0].imag - 0.3,
        f'ŒîRe = {v[0].real:+.0f}', ha='center', fontsize=8, color='gray')
ax.text(w[0].real + 0.15, (p[0].imag + w[0].imag)/2,
        f'Œîim = {v[0].imag:+.0f}', ha='left', fontsize=8, color='gray')

mid = ((p[0] + w[0]) / 2)
ax.text(mid.real - 0.5, mid.imag + 0.2,
        f'|v‚ÇÅ| = {abs(v[0]):.3f}', fontsize=8, color='black')

ax.axhline(0, color='k', linewidth=0.5)
ax.axvline(0, color='k', linewidth=0.5)
ax.set_xlabel('Re'); ax.set_ylabel('Im')
ax.grid(True, linestyle='--', alpha=0.4)
ax.set_aspect('equal')

# ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
# Plot 2: Component 2  (p[1], w[1])
# ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
ax = axes[1]
ax.set_title('Component 2', fontsize=11)

pts2 = {'p‚ÇÇ = 3-2i': p[1], 'w‚ÇÇ = 1+i': w[1]}
colors2 = {'p‚ÇÇ = 3-2i': 'royalblue', 'w‚ÇÇ = 1+i': 'tomato'}

for label, z in pts2.items():
    ax.scatter(z.real, z.imag, s=120, color=colors2[label], zorder=5)
    ax.annotate(label, (z.real, z.imag),
                textcoords='offset points', xytext=(8, 5), fontsize=9, color=colors2[label])

ax.annotate('', xy=(w[1].real, w[1].imag), xytext=(p[1].real, p[1].imag),
            arrowprops=dict(arrowstyle='->', color='black', lw=1.5))

ax.plot([p[1].real, w[1].real], [p[1].imag, p[1].imag],
        color='gray', linestyle=':', linewidth=1)
ax.plot([w[1].real, w[1].real], [p[1].imag, w[1].imag],
        color='gray', linestyle=':', linewidth=1)

ax.text((p[1].real + w[1].real)/2, p[1].imag - 0.3,
        f'Œîre = {v[1].real:+.0f}', ha='center', fontsize=8, color='gray')
ax.text(w[1].real + 0.15, (p[1].imag + w[1].imag)/2,
        f'Œîim = {v[1].imag:+.0f}', ha='left', fontsize=8, color='gray')

mid2 = ((p[1] + w[1]) / 2)
ax.text(mid2.real + 0.1, mid2.imag + 0.2,
        f'|v‚ÇÇ| = {abs(v[1]):.3f}', fontsize=8, color='black')

ax.axhline(0, color='k', linewidth=0.5)
ax.axvline(0, color='k', linewidth=0.5)
ax.set_xlabel('Re'); ax.set_ylabel('Im')
ax.grid(True, linestyle='--', alpha=0.4)
ax.set_aspect('equal')

# ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
# Plot 3: Distance bar chart
# ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
ax = axes[2]
ax.set_title('Hermitian Norm Breakdown', fontsize=11)

labels  = ['|v‚ÇÅ|¬≤', '|v‚ÇÇ|¬≤', '‚Äñv‚Äñ¬≤ = sum', '‚Äñv‚Äñ']
values  = [abs(v[0])**2, abs(v[1])**2, dist**2, dist]
colors3 = ['steelblue', 'steelblue', 'orange', 'tomato']

bars = ax.bar(labels, values, color=colors3, edgecolor='white', width=0.5)
for bar, val in zip(bars, values):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.3,
            f'{val:.2f}', ha='center', fontsize=9)

ax.set_ylabel('Value')
ax.set_ylim(0, max(values) * 1.2)
ax.grid(True, axis='y', linestyle='--', alpha=0.4)

plt.tight_layout()
plt.show()
`````

`````python
print("Step 1: Compute the difference (subtract p from w)")
print(f"  v = w - p = {v}")
print()

print("Step 2: Compute the squared length of each component")
print(f"  |v‚ÇÅ|¬≤ = {abs(v[0])**2:.1f}   ‚Üí (real part¬≤ + imaginary part¬≤) of component 1")
print(f"  |v‚ÇÇ|¬≤ = {abs(v[1])**2:.1f}   ‚Üí (real part¬≤ + imaginary part¬≤) of component 2")
print()

print("Step 3: Add them together")
print(f"  ‚Äñv‚Äñ¬≤ = {dist**2:.1f}   ‚Üí sum of |v‚ÇÅ|¬≤ + |v‚ÇÇ|¬≤")
print()

print("Step 4: Take the square root to get the actual distance")
print(f"  ‚Äñv‚Äñ = ‚àö{dist**2:.0f} ‚âà {dist:.4f}")
`````
